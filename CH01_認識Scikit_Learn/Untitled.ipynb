## Ch01 認識Scikit_learn(Scipy Toolkit)
<H3>(1) classification分類 (2)regression[rɪˋgrɛʃən]回歸(3)clustering 分群</H3>

<H3>(2) 三大功能:</H3>
<b>1.Estimator[ˋɛstə͵metɚ]估計:Supervised & Unsupervised Learning :classification, regression , clustering 都支援Estimator的** fit & transform function **(核心的演算法)</b><br>
<b>2.Predictor[prɪˋdɪktɚ] :支援predict function，搭配Estimato 機器學習模型&資料預測結果，支援score函式計算預測的accuracy</b><br>
<b>3.Transformer: 資料處理 : Feature Selection, Feature Extraction，Dimensionality reduction</b>

<H3>(3) Scikit-Learn 套件家族</H3>
<b>1.Scikit-Learn 套件: Machine Learning 相關功能</b><br>
<b>2.Scikit-Image 套件: image Segmention, Geometric[dʒɪəˋmɛtrɪk] Transformation， color Space</b><br>
<b>3.Scikit-Optimize套件:支援 Sequential model Optimize相關功能</b><br>
ps: 沒有深度學習

<H3> (4) 課程內容(機器學習與深度學習流程):</H3>
<b>1.收集訓練資料->2.資料處理-> 3.分割資料-> 4.特徵工程-> 5.機器學習-> 6.結果輸出</b><br>
